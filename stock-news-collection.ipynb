```python
stock1 = Vnstock().stock(source='VCI')
```

```python
df_cat = stock1.listing.symbols_by_industries()
df_cat.head()
```

```python
df_cat.shape
```

```python
com_list = df_cat['symbol']
```

```python
def DataCollection(name):
    company = Vnstock().stock(symbol=name, source='TCBS').company
    data = company.news()
    data_concat = company.overview()
    data['symbol'] = name
    data_concat['symbol'] = name
    df_final = pd.merge(data, data_concat[['symbol','industry']], left_on='symbol', right_on='symbol', how='inner')
    return df_final
```

```python
DataCollection('VCB')
```

```python
final = pd.DataFrame([])
prog = 0
for company in com_list:
    print("Number of stock pull: ",prog)
    prog += 1
    try:
        temp = DataCollection(company)
        final = pd.concat([final, temp], ignore_index=True)
    except:
        print('error')
```

```python
final.head()
```

```python
final.to_csv("stock_news.csv", sep='\t', encoding='utf-8', index=False, header=True)
```
